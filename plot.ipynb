```python
df1 = pd.read_csv(filepath_or_buffer="score_history_cramped_room.csv",header=None)
df2 = pd.read_csv(filepath_or_buffer="score_history_asymmetric_advantages_maddpg2.csv",header=None)
df3 = pd.read_csv(filepath_or_buffer="score_history_coordination_ring.csv",header=None)
```

```python
ax = df2.plot(x=0,y=1)
ax.set_title(f'traning MADDPG based agents on Layout2')
ax.set(xlabel='episode', ylabel='Number of Soups made')
ax.get_legend().remove()
```

```python
ax.figure.savefig('fig2_DQN_layout1.png')
```

```python
COLORS = ['tab:blue','tab:green','tab:cyan','tab:purple','tab:brown','grey']#'tab:blue',
```

```python
episode=[df1.shape[0],df2.shape[0],df3.shape[0]]
```

```python
fig, axs = plt.subplots(1, 3,figsize=(5*3, 5))
axs[0].plot(df1.iloc[:,0],df1.iloc[:,1],COLORS[0])

axs[1].plot(df2.iloc[:,0],df2.iloc[:,1],COLORS[1])

axs[2].plot(df3.iloc[:,0],df3.iloc[:,1],COLORS[2])

for i in range(3):
    axs[i].hlines(y=7, xmin=0,xmax=10+episode[i],colors='red', linestyles='--', lw=2, label='targetScore')
    axs[i].set_title(f'Layout{i+1}')

for ax in axs.flat:
        ax.set(xlabel='episode', ylabel='Soups Delivered')
fig.savefig(f'traings.png',bbox_inches='tight')
```
